```python
# Student Record Management System

# List to store student records
students = []

# Function to add a student
def add_student():
    student_id = input("Enter Student ID: ")
    name = input("Enter Name: ")
    age = int(input("Enter Age: "))
    courses = input("Enter courses (comma separated): ").split(",")

    # Remove extra spaces from course names
    courses = [course.strip() for course in courses]

    student = {
        "id": student_id,
        "name": name,
        "age": age,
        "courses": courses
    }

    students.append(student)
    print("Student added successfully!\n")


# Function to view all students
def view_students():
    if not students:
        print("No student records found.\n")
        return

    print("\nStudent Records:")
    for student in students:
        print(f"ID: {student['id']}, Name: {student['name']}, Age: {student['age']}, Courses: {student['courses']}")
    print()


# Function to search student by ID
def search_student():
    search_id = input("Enter Student ID to search: ")

    for student in students:
        if student["id"] == search_id:
            print("Student Found:")
            print(student)
            return

    print("Student not found.\n")


# Function to get unique courses
def get_unique_courses():
    unique_courses = set()

    for student in students:
        unique_courses.update(student["courses"])

    print("Unique Courses:", unique_courses, "\n")


# Function to remove student by ID
def remove_student():
    remove_id = input("Enter Student ID to remove: ")

    for student in students:
        if student["id"] == remove_id:
            students.remove(student)
            print("Student removed successfully!\n")
            return

    print("Student not found.\n")


# Menu-driven program
while True:
    print("===== Student Record Management System =====")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Show Unique Courses")
    print("5. Remove Student")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")

    if choice == "1":
        add_student()
    elif choice == "2":
        view_students()
    elif choice == "3":
        search_student()
    elif choice == "4":
        get_unique_courses()
    elif choice == "5":
        remove_student()
    elif choice == "6":
        print("Exiting program...")
        break
    else:
        print("Invalid choice. Please try again.\n")
```
